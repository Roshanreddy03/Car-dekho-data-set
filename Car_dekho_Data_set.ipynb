Car dekho data set


```python
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

conn = sqlite3.connect('Load_dataset.db')

df = pd.read_csv('cardekho_dataset.csv')
df.to_sql('cardekho_dataset.csv',conn, if_exists='replace',index=False)
print("Table created !")
df
```

Here we don't have any null values. Now we are going to make graphs.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Bar plot

```python
sns.countplot(data=df, x='fuel_type', hue='fuel_type',palette='Set2', legend=False)
plt.title('Number of Cars Available by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count of Cars')
plt.show()
```

```python
sns.countplot(data=df, x='transmission_type',color=("skyblue") )
plt.title('Number of Cars Available by Transmission Type')
plt.xlabel('Transmission')
plt.ylabel('Count of Cars')
plt.show()
```

Histogram

```python
# Distribution of Selling Price
sns.histplot(data=df, x='selling_price', kde=True, bins=30)
plt.title('Distribution of Car Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

Bar chart

```python
# Average Selling Price by Fuel Type
sns.barplot(data=df, x='fuel_type', y='selling_price', hue='fuel_type',palette='Set2', legend=False)
plt.title('Average Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.show()
```

Scatter Plot

```python
# Kilometers Driven vs. Selling Price, color-coded by Transmission
sns.scatterplot(data=df, x='km_driven', y='selling_price', hue='transmission_type')
plt.title('Impact of Kilometers Driven on Selling Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price')
plt.show()
```
